```python
import pandas as pd
df = pd.read_csv("salaries_clean.csv")
df
```

```python
# Create a dataframe with only the categorical variables in the salary data

A = df.select_dtypes(include= ['object','category'])
B = df.select_dtypes(include= 'number')
print("Categorical Dataframe")
display(A.head())
print("\n\nNumerical Dataframe")
display(B.head()) 
```

```python
num = df.select_dtypes(include= 'number')
print("\n\nNumerical Dataframe")
display(num.head()) 
```

```python
# -- Create a dataframe with only the following job title: 'software designer', 'junior software engineer'.

A= df[df['job_title'].isin(['software designer', 'junior software engineer'])]
A
```

```python
# -- Create a dataframe showing only the job title and the salary of all 'Software Developers' in california and Oklahoma.

B= [['job_title', 'annual_base_pay', 'location_state']]
B = df[(df['job_title'] == 'software developer') & (df['location_state'].isin(['CA', 'OK']))]
B
```

```python
# -- Who is the highest paid : 'ceo' or 'senior backend engineer'?

a = df[['annual_base_pay', 'job_title']]
a = a.loc[a.job_title.isin(['ceo', 'senior backend enginee'])]
a.groupby("job_title")[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False)

```

```python
#  what job_title_category will advice me to venture into in the US

D = df[['location_country', 'job_title_category', 'job_title','annual_base_pay']]
D = df.loc[(df["location_country"] == 'US')]
D.groupby(['location_country','job_title_category', 'job_title'])[['annual_base_pay']].mean().sort_values(by= 'annual_base_pay', ascending= False).head(1)

```

```python
# - What is the mean average of 'senior backend engineer' salary
E = df[['job_title','annual_base_pay' ]]
E = df.loc[df['job_title'] == 'senior backend engineer']
E.groupby('job_title')['annual_base_pay'].mean()


```

```python
# -- Create a dataframe of data workers who earns an average salary within 90,000 and 50,000 in the US

F = df[["job_title_category", "annual_base_pay"]]
F = df.loc[(df["job_title_category"] == "Data") &   (df["annual_base_pay"].between(50000, 90000))]
F
```

```python
# -- Which country has the most paid Data job title 

G = df[['location_country', 'job_title_category', 'annual_base_pay', 'job_title']]
G = G.loc[df["job_title_category"] == 'Data']
G.groupby(['location_country', 'job_title_category', 'job_title'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False).head(1)

```

```python
# -- What is the job_title name is the least paid in the US.
H = df[['job_title', 'location_country', 'annual_base_pay']]
H = H.loc[df["location_country"]== 'US']
H.groupby(['location_country',  'job_title'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = True).head(1)

```

```python
# -- Create a dataframe of job_title_category who earns an average salary between 70,000 and 120,000 in the united states

I = df[["job_title_category", "annual_base_pay", "location_country"]]
I = df.loc[(df["location_country"] == "US") &   (df["annual_base_pay"].between(70000, 120000))]
I
```

```python
# -- Create a Create a dataframe of job_title_category who earns an average salary between 120,000 and and 200,000, and show the employer name of the highest paying company.

I = df[["job_title_category", "annual_base_pay", "employer_name"]]
I = df.loc[df["annual_base_pay"].between(120000, 200000)]
I[["job_title_category", "annual_base_pay", "employer_name"]]
```

```python
# -- Data or Software job title Category, who is most paid in the US? Find the mean pay of Data job 
J = df[['job_title_category', 'location_country', 'annual_base_pay']]
J = J.loc[((df['job_title_category'] == "Data") | (df['job_title_category'] == "Software")) & (df['location_country'] == "US")]
J.groupby(['location_country',  'job_title_category'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False)

```

```python
# -- Which job title name - inside the Data job title category, is the most paid in the US.
K = df[['job_title', 'location_country','job_title_category', 'annual_base_pay']]
K = K.loc[(df["location_country"]== 'US') & (df['job_title_category'] == "Data")]
K.groupby(['job_title', 'location_country', 'job_title_category'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False).head(1)
```

```python
# -- Which country has the highest paid Software job title category.
L = df[['job_title', 'location_country','job_title_category', 'annual_base_pay']]
L = L.loc[df['job_title_category'] == "Software"]
L.groupby(['job_title', 'location_country', 'job_title_category'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False)
```

```python
# -- Create a dataframe that shows the location, job_title_category, and the employer name of: 'front-end developer', 'sr. engineering manager', 'staff software engineer', and 'senior data analyst'.
A = df[['job_title', 'job_title_category', 'location_country', 'employer_name']]
A = A.loc[df.job_title.isin(['front-end developer', 'sr. engineering manager', 'staff software engineer','senior data analyst'])]
A = A.groupby(['job_title','location_country', 'job_title_category', 'employer_name']).size().reset_index()
A[["job_title_category","employer_name"]]
```
